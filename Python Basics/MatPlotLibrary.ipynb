```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
```

```python
y = x**2
```

```python
#Functional Method
plt.plot(x,y)
```

```python
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Title')
```

```python
fig = plt.figure()
fig.add_axes([0.7,0.5,0.5,0.7])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([.6,.7,.8,.9])
axes2 = fig.add_axes([0.7,1.0,.3,.4])
axes1.plot(x,y,'r--')
axes1.set_title('BIG')

axes2.plot(y,x,'b')
axes2.set_title('SMALL')
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2)
```

```python
axes
```

```python
for ax in axes:
    print(axes)
```

```python
axes[0][0].plot(x,y,'r')
axes[0][1].plot(y,x,'b')
axes[1][0].plot(y,x,'g')
axes[1][1].plot(x,y,'y')
fig
```

```python
for row in axes:
    for ax in row:
        ax.plot(x, y, 'g')
        ax.set_xlabel('x')
        ax.set_ylabel('y')

fig    
```

```python
x = np.linspace(0,5,11)
fig, axes = plt.subplots(1,2,figsize=(10,4))

axes[0].plot(x,x**2,marker='o',ls='-',lw=10)
axes[0].set_title('Normal')

axes[1].plot(x,x**3)
axes[1].set_yscale('log')
axes[1].set_title('Log')
```
